```python
# Apenas para os casos com mais de 1 compartilhamento
df_virais = df_agrupado[df_agrupado['qtd_compartilhamentos'] > 1]

# Obter percentis ajustados para os virais
p80 = df_virais['qtd_compartilhamentos'].quantile(0.80)
p95 = df_virais['qtd_compartilhamentos'].quantile(0.95)

def classificar_viralidade(qtd):
    if qtd == 1:
        return 0  # Baixíssima viralidade (classe majoritária)
    elif qtd <= p80:
        return 1  # Média
    elif qtd <= p95:
        return 2  # Alta
    else:
        return 3  # Muito alta

df_agrupado['viralidade_nivel'] = df_agrupado['qtd_compartilhamentos'].apply(classificar_viralidade)
```

Classe para transformar textos usando Word2Vec

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import mean_squared_error
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.sparse import hstack, issparse
from gensim.models import Word2Vec

# Classe para transformar textos usando Word2Vec
class Word2VecTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, size=100, window=5, min_count=1, workers=4):
        self.size = size
        self.window = window
        self.min_count = min_count
        self.workers = workers

    def fit(self, X, y=None):
        self.w2v_model = Word2Vec(
            sentences=[text.split() for text in X],
            vector_size=self.size,
            window=self.window,
            min_count=self.min_count,
            workers=self.workers
        )
        return self

    def transform(self, X):
        return np.array([
            np.mean(
                [self.w2v_model.wv[word] for word in text.split() if word in self.w2v_model.wv] or
                [np.zeros(self.size)],
                axis=0
            )
            for text in X
        ])
```

Treinamento, Validação e Teste: Outer K-Fold Cross-Validation;

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import KFold
from scipy.sparse import issparse, hstack
import numpy as np

def evaluate_classifier_with_outer_cv(
    X_train_val, y_train_val, X_test, y_test,
    vectorizer, model, scaler=None, normalizer=None,
    texto_coluna='text_content_anonymous',
    colunas_numericas=None,
    colunas_continuas=None,
    outer_splits=5,
    average='weighted'
):
    outer_cv = KFold(n_splits=outer_splits, shuffle=True, random_state=42)
    results_val = []

    for train_index, val_index in outer_cv.split(X_train_val):
        X_train = X_train_val.iloc[train_index]
        X_val = X_train_val.iloc[val_index]
        y_train = y_train_val.iloc[train_index]
        y_val = y_train_val.iloc[val_index]

        # Vetorização do texto
        X_train_text_vec = vectorizer.fit_transform(X_train[texto_coluna])
        X_val_text_vec = vectorizer.transform(X_val[texto_coluna])
        is_sparse = issparse(X_train_text_vec)

        if normalizer and not is_sparse:
            X_train_text_vec = normalizer.fit_transform(X_train_text_vec)
            X_val_text_vec = normalizer.transform(X_val_text_vec)

        if colunas_numericas:
            X_train_num = X_train[colunas_numericas].copy()
            X_val_num = X_val[colunas_numericas].copy()

            if scaler and colunas_continuas:
                scaler.fit(X_train_num[colunas_continuas])
                X_train_num[colunas_continuas] = scaler.transform(X_train_num[colunas_continuas])
                X_val_num[colunas_continuas] = scaler.transform(X_val_num[colunas_continuas])

            if is_sparse:
                X_train_final = hstack([X_train_text_vec, X_train_num.values])
                X_val_final = hstack([X_val_text_vec, X_val_num.values])
            else:
                X_train_final = np.hstack([X_train_text_vec, X_train_num.values])
                X_val_final = np.hstack([X_val_text_vec, X_val_num.values])
        else:
            X_train_final = X_train_text_vec if is_sparse else np.array(X_train_text_vec)
            X_val_final = X_val_text_vec if is_sparse else np.array(X_val_text_vec)

        model.fit(X_train_final, y_train)
        y_val_pred = model.predict(X_val_final)

        acc = accuracy_score(y_val, y_val_pred)
        prec = precision_score(y_val, y_val_pred, average=average, zero_division=0)
        rec = recall_score(y_val, y_val_pred, average=average, zero_division=0)
        f1 = f1_score(y_val, y_val_pred, average=average, zero_division=0)
        results_val.append((acc, prec, rec, f1))

    # Avaliação final no conjunto de teste
    X_train_val_text_vec = vectorizer.fit_transform(X_train_val[texto_coluna])
    X_test_text_vec = vectorizer.transform(X_test[texto_coluna])
    is_sparse = issparse(X_train_val_text_vec)

    if normalizer and not is_sparse:
        X_train_val_text_vec = normalizer.fit_transform(X_train_val_text_vec)
        X_test_text_vec = normalizer.transform(X_test_text_vec)

    if colunas_numericas:
        X_train_val_num = X_train_val[colunas_numericas].copy()
        X_test_num = X_test[colunas_numericas].copy()

        if scaler and colunas_continuas:
            scaler.fit(X_train_val_num[colunas_continuas])
            X_train_val_num[colunas_continuas] = scaler.transform(X_train_val_num[colunas_continuas])
            X_test_num[colunas_continuas] = scaler.transform(X_test_num[colunas_continuas])

        if is_sparse:
            X_train_val_final = hstack([X_train_val_text_vec, X_train_val_num.values])
            X_test_final = hstack([X_test_text_vec, X_test_num.values])
        else:
            X_train_val_final = np.hstack([X_train_val_text_vec, X_train_val_num.values])
            X_test_final = np.hstack([X_test_text_vec, X_test_num.values])
    else:
        X_train_val_final = X_train_val_text_vec if is_sparse else np.array(X_train_val_text_vec)
        X_test_final = X_test_text_vec if is_sparse else np.array(X_test_text_vec)

    model.fit(X_train_val_final, y_train_val)
    y_test_pred = model.predict(X_test_final)

    test_acc = accuracy_score(y_test, y_test_pred)
    test_prec = precision_score(y_test, y_test_pred, average=average, zero_division=0)
    test_rec = recall_score(y_test, y_test_pred, average=average, zero_division=0)
    test_f1 = f1_score(y_test, y_test_pred, average=average, zero_division=0)

    return (
        np.mean([r[0] for r in results_val]),  # Val Accuracy
        np.mean([r[1] for r in results_val]),  # Val Precision
        np.mean([r[2] for r in results_val]),  # Val Recall
        np.mean([r[3] for r in results_val]),  # Val F1
        test_acc, test_prec, test_rec, test_f1
    )
```

- Função para fazer o pré processamento do texto

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
from unidecode import unidecode

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('portuguese'))

def preprocess_text(text):
    def substituir_dominios(texto):
        def extrair_dominio(url):
            dominio = re.sub(r'^https?://(?:www\.)?|www\.', '', url)
            dominio = re.split(r'[/?#]', dominio)[0]
            return dominio.split('.')[0]
        return re.sub(r'https?://(?:www\.)?\S+|www\.\S+', lambda m: extrair_dominio(m.group(0)), texto)

    # Substituir domínios
    text = substituir_dominios(text)

    # Minúsculas e remoção de acentos
    text = unidecode(text.lower())

    # Remover pontuações
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Substituir emojis e risadas repetidas
    text = re.sub(r'([\U00010000-\U0010FFFF])\1+', r'\1', text)
    text = re.sub(r'(ha){2,}', 'haha', text, flags=re.IGNORECASE)
    text = re.sub(r'(ka){2,}', 'kaka', text, flags=re.IGNORECASE)
    text = re.sub(r'k{2,}', 'kk', text, flags=re.IGNORECASE)

    # Reduzir espaços
    text = re.sub(r'\s+', ' ', text).strip()

    # Remover stopwords
    palavras_filtradas = [word for word in text.split() if word not in stop_words]
    text = ' '.join(palavras_filtradas)

    return text
```
